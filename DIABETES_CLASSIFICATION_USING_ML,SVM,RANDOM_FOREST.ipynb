```python
x=df.drop('Outcome',axis=1)
y=df['Outcome']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=95)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
tap=DecisionTreeClassifier(criterion='entropy')
tap.fit(x_train,y_train)
```

```python
y_pred=tap.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
from sklearn.tree import export_graphviz
import graphviz
graphviz.Source(export_graphviz(tap,feature_names=x.columns))
```

```python
from sklearn.svm import SVC
bad=SVC(kernel='linear')
bad.fit(x_train,y_train)
```

```python
y_pred=bad.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rap=RandomForestClassifier(max_depth=7 ,n_estimators=120,n_jobs=1,random_state=50,oob_score=False)
rap.fit(x_train,y_train)
```

```python
y_pred=rap.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
from sklearn.linear_model import LogisticRegression
vip=LogisticRegression()
vip.fit(x_train,y_train)
```

```python
y_pred=vip.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
rip=KNeighborsClassifier(n_neighbors=17)
rip.fit(x_train,y_train)
```

```python
y_pred=rip.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```
