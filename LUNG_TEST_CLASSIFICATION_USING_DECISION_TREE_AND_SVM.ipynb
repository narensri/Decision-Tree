```python
from sklearn.preprocessing import LabelEncoder
rip=LabelEncoder()
df['GENDER']=rip.fit_transform(df['GENDER'])
```

```python
from sklearn.preprocessing import LabelEncoder
rip=LabelEncoder()
df['LUNG_CANCER']=rip.fit_transform(df['LUNG_CANCER'])
```

```python
q=df.corr()
sns.heatmap(q,annot=True)
```

```python
x=df.drop('LUNG_CANCER',axis=1)
y=df['LUNG_CANCER']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=85)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf=DecisionTreeClassifier(criterion='entropy')
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
from sklearn.tree import export_graphviz
import graphviz
graphviz.Source(export_graphviz(clf,feature_names=x.columns))
```

```python
from sklearn.svm import SVC
dip=SVC(kernel='poly')
dip.fit(x_train,y_train)
```

```python
y_pred=dip.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```
