```python
from sklearn.preprocessing import LabelEncoder
tap=LabelEncoder()
df['Crop']=tap.fit_transform(df['Crop'])
```

```python
u=df.corr()
sns.heatmap(u,annot=True)
```

```python
df=df.dropna()
```

```python
x=df.drop(['Crop'],axis=1)
y=df['Crop']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=95)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
rap=DecisionTreeClassifier(criterion='entropy')
rap.fit(x_train,y_train)
```

```python
y_pred=rap.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
from sklearn.svm import SVC
wat=SVC(kernel='linear')
wat.fit(x_train,y_train)
```

```python
y_pred=wat.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
red=KNeighborsClassifier(n_neighbors=7)
red.fit(x_train,y_train)
```

```python
y_pred=red.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
wed=RandomForestClassifier(n_estimators=100,n_jobs=1,max_depth=7,random_state=50,oob_score=True,criterion='entropy')
wed.fit(x_train,y_train)
```

```python
y_pred=wed.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
from sklearn.linear_model import LogisticRegression
zat=LogisticRegression()
zat.fit(x_train,y_train)
```

```python
y_pred=zat.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
from sklearn.naive_bayes import GaussianNB
rip=GaussianNB()
rip.fit(x_train,y_train)
```

```python
y_pred=rip.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```
