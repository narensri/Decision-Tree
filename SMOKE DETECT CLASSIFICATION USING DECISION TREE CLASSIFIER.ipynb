```python
x=df.drop('Fire Alarm',axis=1)
y=df['Fire Alarm']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=95)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
tap=DecisionTreeClassifier(criterion='entropy')
tap.fit(x_train,y_train)
```

```python
y_pred=tap.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
from sklearn.tree import export_graphviz
import graphviz
graphviz.Source(export_graphviz(tap,feature_names=x.columns))
```
