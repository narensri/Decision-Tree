```python
from sklearn.preprocessing import LabelEncoder
rip=LabelEncoder()
df['diagnosis']=rip.fit_transform(df['diagnosis'])
```

```python
n=df.corr()
sns.heatmap(n,annot=True)
```

```python
x=df.drop('diagnosis',axis=1)
y=df['diagnosis']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.15,random_state=95)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier(criterion='entropy')
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
from sklearn.tree import export_graphviz
import graphviz
graphviz.Source(export_graphviz(clf,feature_names=x.columns))
```
