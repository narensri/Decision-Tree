```python
df=df.dropna()
```

```python
x=df.drop(['Survived'],axis=1)
y=df['Survived']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=95)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
bit=DecisionTreeClassifier(criterion='entropy')
bit.fit(x_train,y_train)
```

```python
y_pred=bit.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
from sklearn.tree import export_graphviz
import graphviz
graphviz.Source(export_graphviz(bit,feature_names=x.columns))
```

```python
from sklearn.svm import SVC
tip=SVC(kernel='linear')
tip.fit(x_train,y_train)
```

```python
y_pred=tip.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```
