```python
from sklearn.preprocessing import LabelEncoder
zip=LabelEncoder()
df['Surname']=zip.fit_transform(df['Surname'])
```

```python
from sklearn.preprocessing import LabelEncoder
zip=LabelEncoder()
df['Gender']=zip.fit_transform(df['Gender'])
```

```python
x=df.drop(['Exited'],axis=1)
y=df['Exited']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=99)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
gig=DecisionTreeClassifier(criterion='entropy')
gig.fit(x_train,y_train)
```

```python
y_pred=gig.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
from sklearn.svm import SVC
dip=SVC(kernel='poly')
dip.fit(x_train,y_train)
```

```python
y_pred=dip.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
red=KNeighborsClassifier(n_neighbors=17)
red.fit(x_train,y_train)
```

```python
y_pred=red.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
war=RandomForestClassifier(max_depth=10,n_estimators=200,n_jobs=1,random_state=100,max_features=14,oob_score=True)
war.fit(x_train,y_train)
```

```python
y_pred=war.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
from sklearn.linear_model import LogisticRegression
rap=LogisticRegression()
rap.fit(x_train,y_train)
```

```python
y_pred=rap.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```
